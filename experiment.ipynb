#### Python operators
Operators are used to perform operations on variables and values.  
Python divides the operators in the following groups:  

Arithmetic operators  
Assignment operators  
Comparison operators  
Logical operators  
Identity operators  
Membership operators  
Bitwise operators  

```python
# Arithmetic operators
print(5 + 3, 5 - 3, 5 * 3, 5 / 2, 5 % 2, 5 ** 2, 5 // 2)

# Assignment operators
x = 10; x += 5; print(x)

# Comparison operators
print(5 == 5, 5 != 3, 5 > 3, 5 < 10, 5 >= 5, 5 <= 10)

# Logical operators
print(True and False, True or False, not True)

# Identity operators
a = [1, 2]; b = a; c = [1, 2]; print(a is b, a is not c)

# Membership operators
print(2 in [1, 2, 3], 4 not in [1, 2, 3])

# Bitwise operators
print(5 & 3, 5 | 3, 5 ^ 3, ~5, 5 << 1, 5 >> 1)
```

## Match
The match statement is used to perform different actions based on different conditions.   
instead of writing many if..else statements, you can use the match statement.

The match statement selects one of many code blocks to be executed.

 

```python
print(''' Syntax
match expression:  
  case x:  
      code block  
  case y:      
    code block   
  case z:     
    code block  ''')
```

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

## While Loops
With the while loop we can execute a set of statements as long as a condition is true.

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
### break statement
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

## For Loops
A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
### Looping through a string
for x in "banana":
  print(x)
```

```python
### using break
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

```python
### Using range function
for x in range(6):
  print(x)
```

```python
### else
for x in range(6):
  if x == 3: break
  print(x)
else:
  print("Finally finished!")
```

```python
### List comprehension
newlist = [x for x in fruits if x != "apple"]
print(newlist)
```

```python
## Write a program that reads a number N and prints two Right Angled Triangles of N rows, using numbers starting from 1.
n = int(input("Enter N: "))

print("First Triangle:")
num = 1
for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()

print("Second Triangle:")
num = 1
for i in range(n, 0, -1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
##Write a program to print the factorial of N. Factorial is the product of all positive integers less than or equal to N.
n = int(input("Enter N: "))
fact = 1
for i in range(1, n + 1):
    fact *= i
print("Factorial of", n, "is:", fact)
```

```python
## Write a program that reads a string and prints the count of vowels in the string.
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for ch in s:
    if ch in vowels:
        count += 1
print(f"Number of vowels in {s}:", count)
```

```python
### Given two numbers X and N, write a program to print the sum of N terms in the series

x = int(input("Enter X: "))
n = int(input("Enter N: "))

total = 0
term_str = ""   

for i in range(1, n + 1):
    term_str += str(x)          
    term = int(term_str) ** 2  
    total += term

print(f"Sum of series of {x} and {n}:", total)
```

```python
### given a string , write a program that prints only alphabets

s = input("Enter a string: ")

result = ""
for ch in s:
    if ch.isalpha():   # check if character is alphabet
        result += ch

print(f"Only alphabets in {s}:", result)
```

```python
## write a program that prints all uppercase letters in a string
s = input("Enter a string: ")

result = ""
for ch in s:
    if ch.isupper():   # checks if character is uppercase (A-Z)
        result += ch

print(f"Uppercase letters in {s}:", result)

```

## Data Preprocessing and EDA

##### 1. Download and preprocess air quality datasets

What it means:  

-- Get raw datasets from sources like Kaggle, UCI ML Repository, or government portals (e.g., CPCB, EPA).

-- Datasets usually contain timestamps, pollutant concentrations (PM2.5, PM10, NO2, SO2, CO, O3), temperature, humidity, etc.

Preprocessing tasks:

-- Handle missing values (e.g., interpolation, forward/backward fill).

-- Convert timestamps to proper datetime format.

-- Normalize or scale values (important for ML models).

-- Remove duplicates or anomalies.

##### 2. Conduct EDA (Exploratory Data Analysis) to understand pollutant trends and correlations

What it means:  
EDA = analyzing the dataset visually and statistically to extract insights.

Tasks include:

-- Trend analysis: Plot pollutant levels over time (daily, monthly, yearly).

-- Seasonal patterns: Check how pollution changes with seasons.

-- Correlation analysis: Use heatmaps to see which pollutants are strongly correlated (e.g., PM2.5 often correlates with PM10).

-- Comparisons: Compare pollution levels across cities or monitoring stations.

-- Anomaly detection: Identify sudden pollution spikes (fires, industrial events).

##### 3. Resample data and prepare features for forecasting

What it means:  
Air quality data is often collected hourly/minutely. Forecasting models usually need data at fixed intervals.

Steps:

1. Resampling: Convert raw data into daily/hourly averages (e.g., df.resample('D').mean() in pandas).

2. Feature engineering: Create new features like:

3. Lag features (pollution levels from previous days/hours).

4. Rolling averages (e.g., 7-day moving average of PM2.5).

5. Weather features (temperature, humidity, wind speed).

6. Time-based features (day of week, month, season).

These engineered features help ML/DL models (like LSTMs, ARIMA, XGBoost) forecast future pollution levels.
